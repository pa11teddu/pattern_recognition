1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

**Question 1: Is matrix T invertible?**

Yes, a permutation matrix is always invertible. This is because the inverse of a permutation matrix is simply the transpose of the matrix. The transpose of a matrix is obtained by swapping the rows and columns of the matrix.

To show that a permutation matrix is invertible, we can consider the following:

* A permutation matrix is a square matrix, meaning that it has the same number of rows and columns.
* A permutation matrix is one-to-one, meaning that each row of the matrix corresponds to a unique row of the original image.
* A permutation matrix is onto, meaning that every row of the original image is represented by a row in the permutation matrix.

These three properties together imply that the permutation matrix is invertible.

**Question 2: How would you show that matrix T is invertible?**

To show that matrix T is invertible, we can calculate the determinant of T. If the determinant of T is nonzero, then the matrix is invertible. Otherwise, the matrix is not invertible.

**This code snippet checks if the permutation matrix `T` is invertible by calculating its determinant.** If the determinant is nonzero, then the matrix is invertible and its inverse can be calculated using the `np.linalg.inv()` function. The inverse of the permutation matrix can then be used to perform the inverse scrambling operation on the transformed image `transformed_image` to retrieve the original image. The recovered original image is then displayed using the `plt.imshow()` function.

If the determinant of the permutation matrix is zero, then the matrix is not invertible and the original image cannot be retrieved. In this case, the code snippet prints a message to the console indicating that the matrix is not invertible.


```python
# Calculate the determinant of T to check if it is invertible.
determinant_T = np.linalg.det(T)

# If the determinant is nonzero, then the matrix is invertible.
if determinant_T != 0:

    # Calculate the inverse of T.
    Tinv = np.linalg.inv(T)

    # Perform the inverse scrambling operation to retrieve the original image.
    recovered_image = Tinv @ transformed_image

    # Plot the recovered original image.
    plt.imshow(recovered_image)

else:

    # If the determinant is zero, then the matrix is not invertible and the original image cannot be retrieved.
    print("Matrix T is not invertible (its determinant is zero).")
```
